```python
import pyodbc
```

```python
class ConexaoBD():
 def conexao(self):
  return pyodbc.connect (driver = '{SQL Server}', server = 'TI08', database = 'empresa', trusted_connection = 'yes')
```

```python
bd = ConexaoBD()
conexao = bd.conexao()
```

```python
class Categoria:
 def __init__ (self, conexao):
  self.conexao = conexao
 def criar (self):
  ddl = '''
   create table Categoria (
   id int primary key identity (1, 1),
   nome varchar (32),
   descricao text
  );
  '''
  with self.conexao:
   self.conexao.execute (ddl)
 
 def ler(self):
  cursor = self.conexao.cursor ()
  cursor.execute ('select * from Categoria')
  for linha in cursor:
   print (linha)
```

```python
categoria = Categoria (conexao)
categoria.criar()
```

```python
categoria = Categoria (conexao)
categoria.ler()
```

```python
class Vendedor:
 def __init__ (self, conexao):
  self.conexao = conexao
 def criar (self):
  ddl = '''
   CREATE TABLE Vendedor (
   cpf varchar (14) primary key,
   nome varchar (128),
   email varchar (128),
   telefone varchar (20),
   endereco varchar (256)
  );
  '''
  with self.conexao:
   self.conexao.execute(ddl)

 def ler(self):
  cursor = self.conexao.cursor ()
  cursor.execute ('select * from Vendedor')
  for linha in cursor:
   print (linha)
```

```python
vendedor = Vendedor(conexao)
vendedor.criar()
```

```python
vendedor = Vendedor(conexao)
vendedor.ler()
```

```python
class Produto:
 def __init__(self, conexao):
  self.conexao = conexao
 def criar(self):
  ddl = '''
   CREATE TABLE Produto (
   id int primary key identity (1, 1),
   nome varchar (128),
   descricao text,
   marca varchar (32),
   id_categoria int foreign key references Categoria (id)
  );
  '''
  with self.conexao:
   self.conexao.execute(ddl)

 def ler(self):
  cursor = self.conexao.cursor ()
  cursor.execute ('select * from Produto')
  for linha in cursor:
   print (linha)
```

```python
produto = Produto(conexao)
produto.criar()
```

```python
produto = Produto(conexao)
produto.ler()
```

```python
class VendedorProduto:
 def __init__(self, conexao):
  self.conexao = conexao
 def criar(self):
  ddl = '''
   CREATE TABLE VendedorProduto (
   id int primary key identity (1, 1),
   cpf_vendedor varchar (14) foreign key references Vendedor (cpf),
   id_produto int foreign key references Produto (id),
   quantidade int,
   preco decimal
  );
  '''
  with self.conexao:
   self.conexao.execute(ddl)

 def ler(self):
  cursor = self.conexao.cursor ()
  cursor.execute ('select * from VendedorProduto')
  for linha in cursor:
   print (linha)
```

```python
vendeProdu = VendedorProduto(conexao)
vendeProdu.criar()
```

```python
vendeProdu = VendedorProduto(conexao)
vendeProdu.ler()
```
