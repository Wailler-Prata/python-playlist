```
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import statistics as st
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier as tree
from sklearn import metrics
from sklearn import svm
```

```
dataSet = pd.read_csv('https://pycourse.s3.amazonaws.com/banknote_authentication.txt', header=None, names=['variance', 'skewness', 'curtosis', 'entropy', 'class'])
```

```
#1.	Qual o tamanho desse dataset (número de linhas, número de colunas)?
dataSet
```

```
#y = pd.read_csv('https://pycourse.s3.amazonaws.com/banknote_authentication.txt', header=None, names=['class'])
#x = pd.read_csv('https://pycourse.s3.amazonaws.com/banknote_authentication.txt', header=None, names=['variance', 'skewness', 'curtosis', 'entropy'])
x = dataSet.values[:,0:5]
y = dataSet.values[:,4]
```

```

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size =0.3, random_state = 1)
```

```
x_train
```

```
y_train
```

```
clf_KNN = KNeighborsClassifier(n_neighbors=5)
```

```
clf_KNN.fit(x_train, y_train)
```

```
clf_KNN.predict(x_test)
```

```
clf_KNN.score(x_train, y_train)
```

```
#3.	Qual a média da coluna entropy?
dataSet['entropy'].mean()
```

```
#6.	Qual a porcentagem de exemplos do dataset são cédulas falsas (class=1)?
dataClass = dataSet['class']
qtd = dataClass[dataClass == 1].count()
print((qtd / dataClass.count())* 100)
```

```
#7.	Qual a acurácia do KNN no conjunto de teste?
ac = []
clf = KNeighborsClassifier()
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
ac.append(accuracy_score(y_test, y_pred))
print(ac)
```

```
#8.	Qual a acurácia da Árvore de Decisão no conjunto de teste?
clf = DecisionTreeClassifier(random_state=1)

clf.fit(x_train,y_train)
y_pre = clf.predict(x_test)

print(metrics.accuracy_score(y_test, y_pre))
```

```
#9.	Qual a acurácia do SVM no conjunto de teste?
clf = svm.SVC(gamma='auto',kernel='rbf', random_state=1)
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
print(metrics.accuracy_score(y_test, y_pred))
```
