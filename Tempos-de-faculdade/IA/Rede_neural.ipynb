```
import numpy as np

# Obter o conjunto de amostras de treinamento
entradas = np.array([[0,0],[0,1], [1,0], [1,1]]) #Entrada problema AND
# Associar a saída desejada para cada amostra obtida
saidas = np.array([0,0,0,1]) #Saída problema AND
# Iniciar o vetor de pesos com valores pequenos
pesos = np.array([0.0, 0.0])
# Especificar a taxa de aprendizagem
taxaAprendizagem = 0.1
```

# Função de ativação

```
# Função de ativação do tipo Degrau

def funcaoDegrau(soma):
  if (soma >= 1):
    return 1
  return 0
```

# Atualização do Pesos (A rede neural aprende os pesos)

```
def encontrarPesos(entrada):
  soma = entrada.dot(pesos) #entrada[i] * peso[i]
  return funcaoDegrau(soma)
```

#Realiza o treinamento da rede para encontrar os pesos corretos

```
def treinamento():
    erroTotal = 1
    while (erroTotal != 0):
        erroTotal = 0
        for i in range(len(saidas)):
            novosPesos = encontrarPesos(np.asarray(entradas[i]))
            erro = saidas[i] - novosPesos
            erroTotal += erro
            for j in range(len(pesos)):
                print('Taxa Aprendizagem: ' + str(taxaAprendizagem))
                #+  + str(erro))
                print('Entrada:' + str(entradas[i][j]))
                print('Erro:' + str(erro))
                pesos[j] = pesos[j] + (taxaAprendizagem * entradas[i][j] * erro)
                print('Peso atualizado: ' + str(pesos[j]))
        print('Total de erros: ' + str(erroTotal))
        
# Executa o treinamento
treinamento()
```
