* Identifique as tendências centrais (média) como idade, peso,
altura, avaliação, potencial dos jogadores

```
mediaIdade = dataset['Age'].mean()
print('A idade média é ', round(mediaIdade))

#mediaPeso = dataset['Weight'].mean()
#print('O peso médio é ', mediaPeso)

#mediaAltura = dataset['Height'].mean()
#print('A Altura média é ', round(mediaAltura))

mediaAvaliacao = dataset['Overall'].mean()
print('A avaliação de desempenho médio é ', round(mediaAvaliacao))

mediaPotencial = dataset['Potential'].mean()
print('O potencial médio é ', round(mediaPotencial))
```

* Identifique Mediana, Quartil, Moda destas características

```
medianaIdade = dataset['Age'].median()
print('A mediana da idade é ', round(medianaIdade))

QuartilIdade = dataset['Age'].quantile()
print('O Quartil da idade é  ', round(QuartilIdade))

#mediaPeso = dataset['Weight'].mean()
#print('O peso médio é ', mediaPeso)

#mediaAltura = dataset['Height'].mean()
#print('A Altura média é ', round(mediaAltura))

medianaAvaliacao = dataset['Overall'].median()
print('A mediana da avaliação de desempenho é ', round(medianaAvaliacao))

QuartilIdade = dataset['Overall'].quantile()
print('O Quartil da avaliação de desempenho é ', round(QuartilIdade))

medianaPotencial = dataset['Potential'].median()
print('A mediana do potencial é ', round(medianaPotencial))


QuartilIdade = dataset['Potential'].quantile()
print('O Quartil do Potential é ', round(QuartilIdade))
```

* Identifique as medidas separatrizes (maior, menor, amplitude,
desvio padrão, variância)

```
print('Statisticas da idade')
print()
maxIdade = dataset['Age'].max()
print('A maior da idade é ', round(maxIdade))

minIdade = dataset['Age'].min()
print('A menor da idade é ', round(minIdade))

amplitudeIdade = maxIdade - minIdade
print('A amplitude da idade é ', amplitudeIdade) 

varianceIdade = dataset['Age'].var()
print('A variancia é ', varianceIdade)

DesvPadraoIdade = dataset['Age'].std()
print('O desvio padrão da idade é ', DesvPadraoIdade)



#mediaPeso = dataset['Weight'].mean()
#print('O peso médio é ', mediaPeso)

#mediaAltura = dataset['Height'].mean()
#print('A Altura média é ', round(mediaAltura))
print()
print('Statisticas de Avaliação')
print()

maxAvali = dataset['Overall'].max()
print('A maior Avaliação é ', round(maxAvali))

minAvali = dataset['Overall'].min()
print('A menor Avaliação é ', round(minAvali))

amplitudeAvali = maxAvali - minAvali
print('A amplitude da Avaliação é ', amplitudeAvali) 

varianceAvali = dataset['Overall'].var()
print('A variancia é ', varianceAvali)

DesvPadraoAvali = dataset['Overall'].std()
print('O desvio padrão da Avaliação é ', DesvPadraoAvali)





print()
print('Statisticas de Potential')
print()

maxPotential = dataset['Potential'].max()
print('A maior Avaliação é ', round(maxPotential))

minPotential = dataset['Potential'].min()
print('A menor Avaliação é ', round(minPotential))

amplitudePotential = maxPotential - minPotential
print('A amplitude da Potential é ', amplitudePotential) 

variancePotential = dataset['Potential'].var()
print('A variancia é ', variancePotential)

DesvPadraoPotential = dataset['Potential'].std()
print('O desvio padrão da Avaliação é ', DesvPadraoPotential)
```

```
jogadoresAge = dataset[dataset['Age'] == 25]
jogadoresAge = jogadoresAge['Age'].count()

jogadoresAvali = dataset[dataset['Overall'] == 66]
jogadoresAvali = jogadoresAvali['Overall'].count()

jogadoresPoten = dataset[dataset['Potential'] == 71]
jogadoresPoten = jogadoresPoten['Potential'].count()

print(' A quantidade de jogadores que estão com a mesma idade da média é ',  jogadoresAge, '\n A quntidade de Jogadores com a mesma avaliação da media é ', jogadoresAvali, '\n A quntidade de Jogadores com a mesma avaliação da media é ', jogadoresPoten)
```
