```
import numpy as np
import pandas as pd
```

```
df = pd.read_csv("https://pycourse.s3.amazonaws.com/bike-sharing.csv")
```

* a.	Qual o tamanho desse dataset?

```
df.info()
# 17379  linhas
# 17 colunas
```

* b.	Qual a média da coluna windspeed?

```
mediaWindspeed = sum(df.windspeed)/ len(df.windspeed)
print(mediaWindspeed)
```

* c.	Qual a média da coluna temp?

```
mediaTemp = sum(df.temp)/ len(df.temp)
print(mediaTemp)
```

* d.	Quantos registros existem para o ano de 2011?

```
ano2011 = df[(df['datetime'] >= '2011-01-01') & (df['datetime'] <= '2011-12-31')]
print(ano2011)
# 8645 
```

* e.	Quantos registros existem para o ano de 2012?

```
ano2012 = df[(df['datetime'] >= '2012-01-01') & (df['datetime'] <='2012-12-31')]
print(ano2012)
# 8734 
```

* f.	Quantas locações de bicicletas foram efetuadas em 2011?

```
data = df[(df['datetime'] >= '2011-01-01') & (df['datetime'] <='2011-12-31')]
totalAlguel2011 = sum(data['total_count'])

print(totalAlguel2011) 
```

* g.	Quantas locações de bicicletas foram efetuadas em 2012?

```
data = df[(df['datetime'] >= '2012-01-01') & (df['datetime'] <='2012-12-31')]
totalAlguel2012 = sum(data['total_count'])

print(totalAlguel2012) 
```

```
df['season'].value_counts().plot.bar(figsize=(10, 5),
                                             rot=0);
# codigo da estação 3 equivale a verão, ela possui a maior locação de bicicletas
```

* i.	Qual estação do ano contém a menor média de locações de bicicletas?

```
df['season'].value_counts().plot.bar(figsize=(10, 5),
                                             rot=0);
# codigo da estação 1 equivale a inverno, ela possui a menor locação de bicicletas
```

* j.	Qual horário do dia contém a maior média de locações de bicicletas?

```
melhorhora = df.groupby(by='hour').mean()
melhorhora.sort_values(by = 'total_count')
# As 17:00h
```

* k.	Qual horário do dia contém a menor média de locações de bicicletas?

```
melhorhora = df.groupby(by='hour').mean()
melhorhora.sort_values(by = 'total_count')
# As 04:00h
```

* l.	Que dia da semana contém a maior média de locações de bicicletas?

```
melhorhora = df.groupby(by='weekday').mean()
melhorhora.sort_values(by = 'total_count')
# No dia 4, equivalente a Quinta feira
```

* m.	Que dia da semana contém a menor média de locações de bicicletas?

```
melhorhora = df.groupby(by='weekday').mean()
melhorhora.sort_values(by = 'total_count')
# No dia 0, equivalente a Domingo
```

* n.	Às quartas-feiras (weekday = 3), qual o horário do dia contém a maior média de locações de bicicletas?

```
quarta = df[(df['weekday'] == 3)]
quarta.groupby(by='hour').mean().sort_values(by = 'total_count')
# A maior utilização na quarta feira é as 17:00h
```

* o.	Aos sábados (weekday = 6), qual o horário do dia contém a maior média de locações de bicicletas?

```
quarta = df[(df['weekday'] == 6)]
quarta.groupby(by='hour').mean().sort_values(by = 'total_count')
# A maior utilização no domingo é as 13:00h
```
