# Conjunto

# Declaração


```
#Criar um conjunto:
conjunto = {0, 1, 67.2, True, False, "Ana Conda", (7, 89.6, True, False
, "Python")}
print (conjunto)
print (type (conjunto))
```

* No exemplo, os valores False e True não são impressos, porque um conjunto não aceita
valores repetidos. Eles são eliminados, porque o conjunto do exemplo possui os valores 0 e
1, que são associados a falso e verdadeiro, respectivamente. Além disso, os conjunto não
são ordenáveis. Assim, não é possível determinar a ordem de impressão dos elementos:

* O construtor "set ()" também pode ser usado para declarar um conjunto com elementos
passados como parâmetro. Os elementos precisam ser passados como parâmetro em um
conjunto, lista ou tupla. Exemplo: execute o trecho de código abaixo no Google Colab. Crie
um campo Markdown em seu notebook para identificar o exemplo.

```
#Criar um conjunto (parâmetros passados em um conjunto):
conjunto = set ({0, 1, 67.2, True, False, "Ana Conda", (7, 89.6, True,
False, "Python")})
print (conjunto)
print (type (conjunto))
```

* ou

```
#Criar um conjunto (parâmetros passados em uma lista):
conjunto = set ([0, 1, 67.2, True, False, "Ana Conda", (7, 89.6, True,
False, "Python")])
print (conjunto)
print (type (conjunto))
```

* ou

```
#Criar um conjunto (parâmetros passados em uma tupla):
conjunto = set ((0, 1, 67.2, True, False, "Ana Conda", (7, 89.6, True,
False, "Python")))
print (conjunto)
print (type (conjunto))
```

Exemplo: execute o trecho de código abaixo no Google Colab. Crie um campo Markdown
em seu notebook para identificar o exemplo

```
#Acesso:
conjunto = {0, 1, 67.2, True, False, "Ana Conda", (7, 89.6, True, False
, "Python")}
for elemento in conjunto:
 print (" %-35s %s" % (str(elemento), type (elemento)))
```

# Existência

* Também é possível verificar se um elemento pertence a um conjunto, usando a palavra
reservada "in". Exemplo:

```
#Acesso:
conjunto = {0, 1, 67.2, True, False, "Ana Conda", (7, 89.6, True, False
, "Python")}
if (7, 89.6, True, False, "Python") in conjunto:
 print ("Ok! O elemento pertence ao conjunto!")
```

# Alteração

* Depois que um conjunto é declarado, não é possível alterar seus elementos. Entretanto, é
possível inserir e remover elementos.

# Inserção

* O método "add ()" é usado para adicionar um único novo elemento a um conjunto.
Exemplo: 

```
#Inserção:
conjunto = {0, 1, 67.2, True, False, "Ana Conda"}
print (conjunto)
conjunto.add ((7, 89.6, True, False, "Python"))
print (conjunto)
```

* O método "update ()" é usado para adicionar mais de um novo elemento a um conjunto.
Exemplo: 

```
#Inserção:
conjunto = {0, 1, 67.2, True, False}
print (conjunto)
conjunto.update (["Ana Conda", (7, 89.6, True, False, "Python")])
print (conjunto)
```

# Remoção

* A remoção de um elemento de um conjunto pode ser realizada através dos métodos
"remove ()" ou "discard ()". Os dois métodos aceitam apenas um parâmetro por vez e não
retornam o elemento removido. O método "remove ()" remove um element do conjunto, caso o conjunto possua o
elemento. Exemplo:

```
#Remoção:
conjunto = {0, 1, 67.2, True, False, "Ana Conda", (7, 89.6, True, False
, "Python")}
print (conjunto)
a = conjunto.remove (True)
print (conjunto)
print (a)
```

* Caso o conjunto não possua o elemento, o método "remove ()" dispara uma mensagem de
erro. Exemplo: execute o trecho de código abaixo no Google Colab. Crie um campo
Markdown em seu notebook para identificar o exemplo. O exemplo tenta remover o string
"Python" do conjunto. Entretanto, o conjunto não possui o string "Python", embora o
conjunto possua uma tupla com o string "Python".

```
#Remoção:
conjunto = {0, 1, 67.2, True, False, "Ana Conda", (7, 89.6, True, False, "Python")}
print (conjunto)
conjunto.remove ("Python")
print (conjunto)
```

* O método " discard ()" remove um element do conjunto, caso o conjunto possua o
elemento. Exemplo:

```
#Remoção:
conjunto = {0, 1, 67.2, True, False, "Ana Conda", (7, 89.6, True, False
, "Python")}
print (conjunto)
a = conjunto.discard (False)
print (conjunto)
print (a)
```

* Caso o conjunto não possua o elemento, o método "discard ()" ignora a remoção e
nenhuma mensagem de erro é disparada. Exemplo:

```
#Remoção:
conjunto = {0, 1, 67.2, True, False, "Ana Conda", (7, 89.6, True, False
, "Python")}
print (conjunto)
conjunto.discard ("Python")
print (conjunto)
```

* O método "pop ()" também pode ser usado para remover um elemento de um cojunto. Esse
método removerá o último elemento do conjunto. Entretanto, é importante lembrar que os
conjuntos não são ordenados. Logo, não se sabe qual elemento será removido. O método
"pop ()" retorna o elemento removido. O método "pop ()" não aceita parâmetro. Exemplo:

```
#Remoção:
conjunto = {0, 1, 67.2, True, False, "Ana Conda", (7, 89.6, True, False
, "Python")}
print (conjunto)
a = conjunto.pop ()
print (conjunto)
print (a)
```

* Caso o conjunto esteja vazio, o método "pop ()" irá disparar uma mensagem de erro.
Exemplo:

```
#Remoção:
conjunto = set ()
print (conjunto)
print (type (conjunto))
a = conjunto.pop ()
print (conjunto)
print (a)
```

* O método "clear ()" remove todos os elemento de um conjunto de uma única vez, tornando
o conjunto um conjunto vazio. O método "clear ()" não retorna os elementos removidos.
Exemplo:

```
#Remoção:
conjunto = {0, 1, 67.2, True, False, "Ana Conda", (7, 89.6, True, False
, "Python")}
print (conjunto)
a = conjunto.clear ()
print (conjunto)
print (a)
```

* A palavra-chave "del" remove o conjunto da memória, exclui o conjunto, o conjunto deixa
de existir. Exemplo: 

```
#Remoção:
conjunto = {0, 1, 67.2, True, False, "Ana Conda", (7, 89.6, True, False, "Python")}
print (conjunto)
del conjunto
print (conjunto)
```

# Tamanho

* A função "len ()" é usada para determinar quantos elementos um conjunto possui. Exemplo:

```
#Tamanho de um conunto:
conjunto = {1, 67.2, True, "Ana Conda", (7, 89.6, True, False, "Python"
)}
print ("Conjunto:", conjunto)
print ("Tamanho:", len (conjunto))
```

# Associação

* Há várias maneiras de associar dois ou mais conjuntos na linguagem Python. O "union ()", por
exemplo, realiza a união de dois conjuntos. Esse método retorna um novo conjunto contendo todos
os elementos de ambos os conjuntos. Elementos duplicados, interseção entre os dois conjuntos, são
inseridos uma única vez. Exemplo:

```
#União:
conjunto1 = {0, 1, 67.2, True, "Ana Conda"}
print (conjunto1)
conjunto2 = {False, "Ana Conda", (7, 89.6, True, False, "Python")}
print (conjunto2)
conjunto = conjunto1.union (conjunto2)
print (conjunto)
print (type (conjunto))
```

* O método "update ()" também pode ser usado para associar conjuntos. Como visto
anteriormente, esse método adiciona mais de um novo elemento a um conjunto. Os
elementos podem ser passados como parâmetro para o método em uma lista, em uma
tupla ou em um outro conjunto. Elementos duplicados não são. Exemplo: 

```
#União – elementos passados como parâmetro em uma lista:
conjunto = {"a", "b" , "c", 0, 1, 67.2, True, "Ana Conda"}
lista = [1, 2, 3, False, "Ana Conda", (7, 89.6, True, False, "Python")]
conjunto.update (lista)
print (conjunto)
print (lista)
```

* ou

```
#União – elementos passados como parâmetro em uma tupla:
conjunto = {"a", "b" , "c", 0, 1, 67.2, True, "Ana Conda"}
tupla = (1, 2, 3, False, "Ana Conda", (7, 89.6, True, False, "Python"))
conjunto.update (tupla)
print (conjunto)
print (tupla)
```

* ou

```
#União – elementos passados como parâmetro em um outro conjunto:
conjunto1 = {"a", "b" , "c", 0, 1, 67.2, True, "Ana Conda"}
conjunto2 = {1, 2, 3, False, "Ana Conda", (7, 89.6, True, False, "Pytho
n")}
conjunto1.update (conjunto2)
print (conjunto1)
print (conjunto2)
```
