# Classe range


```
print (type (range (3)))
print("Lista vazia:")
# Inicia uma lista vazia:
print ("Lista vazia:")
lista = list (range (0))
print (lista)
print(" ")

print("Lista em ordem crescente:")
# Inicia uma lista com os inteiros de 0 a 9:
print ("Lista com os inteiros de 0 a 9:")
lista = list (range (10))
print (lista)
print(" ")

print("Lista com definição de início e fim:")
# Inicia uma lista com os inteiros de 5 a 9:
print ("Lista com os inteiros de 5 a 9:")
lista = list (range (5, 10))
print (lista)
print(" ")

print("Lista com definição de incremento:")
# Inicia uma lista com os inteiros de -10 a 8 de 2 em 2:
print ("Lista com os inteiros de -10 a 8 de 2 em 2:")
lista = list (range (-10, 10, 2))
print (lista)
print(" ")

print("Lista em ordem decrescente com definição de decremento:")
# Inicia uma lista com os inteiros de 10 a -8 de -2 em -2:
print ("Lista com os inteiros de 10 a -8 de -2 em -2:")
print (list (range (10, -10, -2)))
print(" ")

print("Concatenação:")
# Concatenação:
print ("Concatenação:")
lista1 = list (range (5))
lista2 = list (range (10, 20, 2))
lista = lista1 + lista2
print (lista)
print(" ")

print("Concatenação usando a função chain ():")
# Concatenação usando a função chain ():
from itertools import chain
print ("Concatenação usando a função chain ():")
concat = chain (range (5), range (10, 20, 2))
lista = list (concat)
print (lista)
print(" ")

print("Valores informados pelo usuário:")
# Valores informados pelo usuário:
inicio = int (input ("Informe o início: "))
fim = int (input ("Informe o fim: "))
passo = int (input ("Informe o incremento|decremento: "))
lista = list (range (inicio, fim, passo))
print (lista)
```

# Operador "Slice"

```
print("Imprimir a lista da posição de índice 4 até a posição de índice 7:")
lista = list (range (11))
print ("Lista completa:", lista)
print ("Parte da lista:", lista[4:8])

print(" ")
print("Imprimir a lista da posição de índice 4 até a posição de índice 7, mas pulando os elementos de 2 em 2:")
lista = list (range (11))
print ("Lista completa:", lista)
print ("Parte da lista:", lista[4:8:2])

print(" ")
print("Imprimir a lista da posição de índice 8 até a posição de índice 4:")
lista = list (range (11))
print ("Lista completa:", lista)
print ("Parte da lista:", lista[8:4:-1])

print(" ")
print("Imprimir a lista da posição de índice 8 até a posição de índice 4, mas pulando os elementos de 2 em 2:")
lista = list (range (11))
print ("Lista completa:", lista)
print ("Parte da lista:", lista[8:4:-2])

print(" ")
print("Imprimir a lista do início ao fim, mas pulando os elementos de 2 em 2:")
lista = list (range (11))
print ("Lista completa:", lista)
print ("Parte da lista:", lista[::2])

print(" ")
print("Imprimir a lista do fim para o início:")
lista = list (range (11))
print ("Lista completa:", lista)
print ("Parte da lista:", lista[::-1])

print(" ")
print("Imprimir uma string do fim para o início:")
str = "Ana Conda"
print ("String:", str)
print ("String invertido:", str[::-1])

print(" ")
print("Inverter lista de strings:")
lista = ["Python", "Ana Conda"]
print ("Lista:", lista)
print ("Lista invertida:", lista[::-1])

print(" ")
print("Imprimir a lista do fim para o início, mas pulando os elementos de 2 em 2:")
lista = list (range (11))
print ("Lista completa:", lista)
print ("Parte da lista:", lista[::-2])

print(" ")
print("Valores informados pelo usuário:")
inicio = int (input ("Informe o início: "))
fim = int (input ("Informe o fim: "))
passo = int (input ("Informe o incremento|decremento: "))
lista = list (range (10))
print ("Lista completa:", lista)
print ("Parte da lista:", lista[inicio:fim:passo])
```

# Acesso aos Elementos de Objeto range

```
#Primeiro elemento:
elemento = range (10)[0]
print("Primeiro elemento:", elemento)

#Segundo elemento:
elemento = range (10)[1]
print("\nSegundo elemento:", elemento)
#Quinto elemento:
elemento = range (10)[4]
print("\nQuinto elemento:", elemento)
#Último elemento:
elemento = range (10)[-1]
print("\nÚltimo elemento:", elemento)
```

# Matriz: Lista de Listas


* Imprimindo através da função "range ()":

```
# Matriz:
matriz = []
matriz.append (list (range (3)))
matriz.append (list (range (5, 10)))
matriz.append (list (range (2, 10, 2)))
print (matriz)
for i in range (len (matriz)):
 for j in range (len (matriz[i])):
  print (matriz[i][j], end = ' ')
 print ()

```

* Imprimindo através das listas em cada linha:

```
# Matriz:
matriz = []
matriz.append (list (range (3)))
matriz.append (list (range (5, 10)))
matriz.append (list (range (2, 10, 2)))
print (matriz)
for linha in matriz:
 for elemento in linha:
  print (elemento, end = ' ')
 print ()
```
