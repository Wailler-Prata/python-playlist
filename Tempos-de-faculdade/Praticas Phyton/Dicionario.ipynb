# Dicionário

* Os dicionário são coleções que não podem ser ordenadas e que não permitem elementos
duplicados. Entretanto, os dicionários podem ser alterados e também podem ser indexados.
Os elementos de um dicionário são formados por pares "chave-valor". Os índices de um
dicionário são as chaves de cada par "chave-valor"

# Declaração

* Um dicionário pode ser declarado através de chaves, {}. Exemplo:

```
#Declaração:
dicionario = {}
print (dicionario)
print (type (dicionario))
```

* Os elementos de um dicionário são compostos por dois campos: uma chave e um valor
relacionado a essa chave. Então, dizemos que cada elemento de um dicionário é formado
por um par chave-valor. Os elementos são separados por vírgulas. Exemplo:

```
#Declaração:
dicionario = {
 "nome": "Ana Conda",
 "saldo": 67.2,
 "carro": 1,
 "moto": 0,
 "graduacao": True,
 "pos-graduacao": False
}
print (dicionario)
print (type (dicionario))
```

 * O construtor "dict ()" também pode ser usado para declarar um dicionário. Exemplo: 

```
#Declaração:
dicionario = dict ()
print (dicionario)
print (type (dicionario))
```

* Na declaração de um dicionário não vazio, os elementos são passados como argumentos
para o construtor "dict ()". Em uma função, procedimento ou método, um parâmetro é uma
variável declarada entre parênteses. Por outro lado, um argumento é o valor passado para o
parâmetro. No construtor "dict ()", os argumentos são passados como pares chave-valor
(keyword arguments, kwargs). Além disso, a ordem em que os argumentos são passados
não é considerada. Exemplo:

```
#Declaração:
dicionario = dict (
 nome = "Ana Conda",
 saldo = 67.2,
 carro = 1,
 moto = 0,
 graduacao = True,
 pos_graduacao = False
)
print (dicionario)
print (type (dicionario))
```

# Acesso

* O acesso aos elementos de um dicionário é realizado atavés do operador "slice", []. A chave
de um par chave-valor tem que ser referenciada dentro do operador. Exemplo:

```
#Acesso:
dicionario = dict (
 nome = "Ana Conda",
 saldo = 67.2,
 carro = 1,
 moto = 0,
 graduacao = True,
 pos_graduacao = False
)
print (dicionario["nome"])
print (type (dicionario["nome"]))
```

* Os dicionários possuem um método "get ()" que tem a mesma funcionalidade do operador
"slice". Exemplo:

```
#Acesso:
dicionario = dict (
 nome = "Ana Conda",
 saldo = 67.2,
 carro = 1,
 moto = 0,
 graduacao = True,
 pos_graduacao = False
)
print (dicionario.get ("nome"))
print (type (dicionario.get ("nome")))
```

* É possível percorrer os elementos de um dicionário, usando uma estrutura de repetição.
Quando percorremos um dicionário empregando uma estrutura de repetição, são
retornados os nomes das chaves do dicionário. Exemplo:

```
#Acesso:
dicionario = dict (
 nome = "Ana Conda",
 saldo = 67.2,
 carro = 1,
 moto = 0,
 graduacao = True,
 pos_graduacao = False
)
for chave in dicionario:
 print (chave, end = " ")
 print (type (chave))
```

* Também podemos usar o método "keys ()" para termos acesso aos nomes das chaves de um
dicionário. Exemplo:

```
#Acesso:
dicionario = dict (
 nome = "Ana Conda",
 saldo = 67.2,
 carro = 1,
 moto = 0,
 graduacao = True,
 pos_graduacao = False
)
for chave in dicionario.keys ():
 print (chave, end = " ")
 print (type (chave))
```

* Para conseguirmos acesso aos valores associados às chaves de um dicionário em uma
estrutura de repetição, empregamos o operador "slice" ou o método "get ()". Exemplo:

```
#Acesso:
dicionario = dict (
 nome = "Ana Conda",
 saldo = 67.2,
 carro = 1,
 moto = 0,
 graduacao = True,
 pos_graduacao = False
)
for chave in dicionario:
 print (dicionario[chave], end = " ")
 print (type (dicionario[chave]))
```

* ou 

```
#Acesso:
dicionario = dict (
 nome = "Ana Conda",
 saldo = 67.2,
 carro = 1,
 moto = 0,
 graduacao = True,
 pos_graduacao = False
)
for chave in dicionario:
 print (dicionario.get (chave), end = " ")
 print (type (dicionario.get (chave)))
```

* Também podemos usar o método "values ()" para termos acesso aos valores associados às
chaves de um dicionário. Exemplo:

```
#Acesso:
dicionario = dict (
 nome = "Ana Conda",
 saldo = 67.2,
 carro = 1,
 moto = 0,
 graduacao = True,
 pos_graduacao = False
)
for valor in dicionario.values ():
 print (valor, end = " ")
 print (type (valor))
```

* O método "items ()" nos permite ter acesso às chaves e seus respectivos valores em um
dicionário. Cada par chave-valor é retornado em uma tupla. Exemplo:

```
#Acesso:
dicionario = dict (
 nome = "Ana Conda",
 saldo = 67.2,
 carro = 1,
 moto = 0,
 graduacao = True,
 pos_graduacao = False
)
for chave, valor in dicionario.items ():
 print (chave, "->", valor)
```

* ou

```
#Acesso:
dicionario = dict (
 nome = "Ana Conda",
 saldo = 67.2,
 carro = 1,
 moto = 0,
 graduacao = True,
 pos_graduacao = False
)
for elemento in dicionario.items ():
 print (elemento, end = " -> ")
 print (type (elemento))
```

# Tamanho

* Usamos a função "len ()" para determinar quantos elementos um dicionário possui. Dessa
forma, o tamanho de um dicionário equivale ao número de pares chave-valor que ele
possui. Exemplo:

```
#Tamanho de um dicionário - número de pares chave-valor:
dicionario = dict (
 nome = "Ana Conda",
 saldo = 67.2,
 carro = 1,
 moto = 0,
 graduacao = True,
 pos_graduacao = False
)
print (dicionario)
print ("Tamanho:", len (dicionario))
```

# Existência

* É possível verificar se uma chave pertence a um dicionário, usando a palavra reservada "in".
Exemplo:

```
#Existência:
dicionario = dict (
 nome = "Ana Conda",
 saldo = 67.2,
 carro = 1,
 moto = 0,
 graduacao = True,
 pos_graduacao = False
)
if "carro" in dicionario:
 print ("O dicionário possui a chave carro.")
```

* O mesmo resultado é alcançado, usando a palavra reservada "in" e o método "keys ()".
Exemplo:

```
#Existência:
dicionario = dict (
 nome = "Ana Conda",
 saldo = 67.2,
 carro = 1,
 moto = 0,
 graduacao = True,
 pos_graduacao = False)
if "carro" in dicionario.keys ():
 print ("O dicionário possui a chave carro.")
```

* Para verificar a existência de um valor, é preciso usar a palavra reservada "in" e o método
"values ()". Exemplo:

```
#Existência:
dicionario = dict (
 nome = "Ana Conda",
 saldo = 67.2,
 carro = 1,
 moto = 0,
 graduacao = True,
 pos_graduacao = False
)
if 67.2 in dicionario.values ():
 print ("O dicionário possui o saldo informado.")
```

* Também é possível verificar a existência de um par chave-valor. Basta empregar a palavra
reservada "in" e o método "items ()". O par chave-valor é passado em uma tupla. Exemplo: 

```
#Existência:
dicionario = dict (
 nome = "Ana Conda",
 saldo = 67.2,
 carro = 1,
 moto = 0,
 graduacao = True,
 pos_graduacao = False
)
if ("nome", "Ana Conda") in dicionario.items ():
 print ("O dicionário possui o item informado.")
```

* ou

```
#Existência:
dicionario = dict (
 nome = "Ana Conda",
 saldo = 67.2,
 carro = 1,
 moto = 0,
 graduacao = True,
 pos_graduacao = False
)
tupla = tuple (("nome", "Ana Conda"))
if tupla in dicionario.items ():
 print ("O dicionário possui o item informado.")
```

# Alteração

* Podemos alterar o valor de um elemento, referenciando sua chave. Exemplo:

```
#Alteração:
dicionario = dict (
 nome = "Ana Conda",
 saldo = 67.2,
 carro = 1,
 moto = 0,
 graduacao = True,
 pos_graduacao = False
)
print (dicionario)
dicionario["nome"] = "Python"
print (dicionario)
```

# Inserção

* Inserimos um novo elemento no dicionário, usando o operador "slice", []. Informamos no
operador "slice" a nova chave como índice e atribuímos a ela o novo valor. Exemplo:

```
#Inserção:
dicionario = dict (
 nome = "Ana Conda",
 saldo = 67.2,
 carro = 1,
 moto = 0,
 graduacao = True,
 pos_graduacao = False
)
print (dicionario)
dicionario["filho"] = "Python"
print (dicionario)
```

* O método "update ()" pode ser usado para inserir elementos em um dicionário. Exemplo: 

```
#Inserção:
dicionario = dict (
 nome = "Ana Conda",
 saldo = 67.2,
 carro = 1,
 moto = 0,
 graduacao = True,
 pos_graduacao = False
)
print (dicionario)
dicionario.update (filho = "Python", casa = 1, telefone = "9 8432 5168")
print (dicionario)
```

* O método "update ()" também pode ser usado para realizar junções entre conjuntos.
Dicionários não aceitam elementos duplicados. Assim, elementos com chaves iguais são
ignorados. Exemplo:

```
#Inserção:
dicionario1 = dict (nome = "Ana Conda",
 saldo = 67.2,
 carro = 1,
 moto = 0,
 graduacao = True,
 pos_graduacao = False
)
dicionario2 = {
 "filho": "Python",
 "casa": 1,
 "telefone": "9 8432 5168"
}
print (dicionario1)
print (dicionario2)
dicionario1.update (dicionario2)
print (dicionario1)
print (dicionario2)
```

# Remoção

* O método "pop ()" remove o elemento cuja chave é informada como argumento. Exemplo: 

```
#Remoção:
dicionario = dict (
 nome = "Ana Conda",
 saldo = 67.2,
 carro = 1,
 moto = 0,
 graduacao = True,
 pos_graduacao = False
)
print (dicionario)
dicionario.pop ("pos_graduacao")
print (dicionario)
```

* O método "popitem ()" remove o último elemento do dicionário. Exemplo:

```
#Remoção:
dicionario = dict (
 nome = "Ana Conda",
 saldo = 67.2,
 carro = 1,
 moto = 0,
 graduacao = True,
 pos_graduacao = False
)
while len (dicionario) > 0:
 print (dicionario)
 dicionario.popitem ()
print (dicionario)
```

* O método "clear ()" remove todos os elemento de um dicionário, tornando o dicionário um
dicionário vazio. O dicionário continua existindo na memória, mas sem elementos. Exemplo:

```
#Remoção:
dicionario = dict (
 nome = "Ana Conda",
 saldo = 67.2,
 carro = 1,
 moto = 0,
 graduacao = True,
 pos_graduacao = False
)
print (dicionario)
dicionario.clear ()
print (dicionario)
```

* A palavra reservada "del" remove o elemento cuja chave é informada no operador "slice",
[]. Exemplo: 

```
#Remoção:
dicionario = dict (
 nome = "Ana Conda",
 saldo = 67.2,
 carro = 1,
 moto = 0,
 graduacao = True,
 pos_graduacao = False
)
print (dicionario)
del dicionario["pos_graduacao"]
print (dicionario)
```

* A palavra reservada "del" também remove o dicionário da memória, exclui o dicionário. O
dicionário deixa de existir. Exemplo: 

```
#Remoção:
dicionario = dict (
 nome = "Ana Conda",
 saldo = 67.2,
 carro = 1,
 moto = 0,
 graduacao = True,
 pos_graduacao = False
)
print (dicionario)
del dicionario
print (dicionario)
```

# Cópia

* Não podemos copiar um dicionário simplesmente fazendo dicionario2 = dicionario 1. Caso
isso seja feito, dicionario2 será apenas uma referência para dicionario 1, e não uma cópia.
Dessa forma, as alterações feitas em dicionario1 também serão automaticamente feitas em
dicionario2, e vice-versa.
Uma maneira de se fazer uma cópia de um dicionário é usar o método "copy ()". Exemplo: 

```
#Cópia:
dicionario1 = dict (
 nome = "Ana Conda",
 saldo = 67.2,
 carro = 1,
 moto = 0,
 graduacao = True,
 pos_graduacao = False
)
dicionario2 = dicionario1.copy ()
print (dicionario1)
print (dicionario2)
```

* O construtor "dict ()" também pode ser usado para criação de cópias. Exemplo:

```
#Cópia:
dicionario1 = dict (
 nome = "Ana Conda",
 saldo = 67.2,
 carro = 1,
 moto = 0,
 graduacao = True,
 pos_graduacao = False
)
dicionario2 = dict (dicionario1)
print (dicionario1)
print (dicionario2)
```

# Dicionários Aninhados

* Um dicionário pode conter vários outros dicionários. Exemplo:

```
#Dicionários Aninhados:
dicionario = {
 "representante": {
 "nome": "Ana Conda",
 "saldo": 67.2,
 "carro": 1,
 "moto": 0,
 "graduacao": True,
 "pos_graduacao": False
 },
 "vendedor": {
 "nome": "Lang Python",
 "saldo": 78.9,
 "carro": 1,
 "moto": 1,
 "graduacao": True,
 "pos_graduacao": True
 },
 "fornecedor": {
 "nome": "Dado Pandas",
 "saldo": 56.4,
 "carro": 1,
 "moto": 0,
 "graduacao": False,
 "pos_graduacao": False
 }
}
print (dicionario)
```

* Também podemos declarar um dicionário formado por outros dicionários já existentes.
Exemplo: 

```
#Dicionários Aninhados:
representante = {
 "nome": "Ana Conda",
 "saldo": 67.2,
 "carro": 1,
 "moto": 0,
 "graduacao": True,
 "pos_graduacao": False
}
vendedor = {
 "nome": "Lang Python",
 "saldo": 78.9,
 "carro": 1,
 "moto": 1,
 "graduacao": True,
 "pos_graduacao": True
}
fornecedor = {
 "nome": "Dado Pandas",
 "saldo": 56.4,
 "carro": 1,
 "moto": 0,
 "graduacao": False,
 "pos_graduacao": False
}
dicionario = {
 "representante": representante,
 "vendedor": vendedor,
 "fornecedor": fornecedor
}
print (dicionario)
```
