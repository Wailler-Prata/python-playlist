* 1) Em um estacionamento há carros e motos. Escreva um algoritmo em linguagem de
programação Python que solicite ao usuário o total de veículos e o total de rodas. O
algoritmo deve determinar quantos carros e quantas motos há no estacionamento. O
resultado deve ser apresentado para o usuário. Acrescente comentários Python em seu
algoritmo.

```python
"""
c = numero de carros
m = numero de motos
v = mumero de viculos > Informados pelo usuário
r = numero de rodas   > Informados pelo usuário

c + m = v
4 * c + 2 * m = r
c = v - m

4 * (c - m) + 2* m = r
4 * v - 4*m + 2 * m = r
4 * v - 2 * m = r
-2 * m = r - 4 * v * (-1)
2 * m = - r + 4 * v
m = ( - r + 4 * v ) / 2
"""
```

```python
nVeiculos = int(input("Informe o numero de veiculos: "))
```

```python
nRodas = int(input("Informe o numero de rodas: "))
```

*  Em linguagem de programação Python, escreva o algoritmo de uma máquina de venda
automática de salgadinhos, doces, sucos e refrigerantes. O algoritmo deve calcular o
menor número de notas que deve ser dado de troco para um pagamento efetuado. O
algoritmo deve ler o valor a ser pago e o valor efetivamente pago. Se o valor
efetivamente pago for menor que o valor a ser pago, a máquina deve apresentar uma
mensagem informando que a quantia paga é insuficiente para realizar a compra. A
máquina aceita apenas notas de R$ 50,00, R$ 20,00, R$ 10,00, R$ 5,00, R$ 2,00 e R$
1,00. Acrescente comentários Python em seu algoritmo.


```python
# Criando um array de notas para serem usadas de troco
notas = [50,20,10,2,1]
```

```python
# Valores de cada iten
salgadinho = 10
doces = 8
sucos = 9
refrigerantes = 5 
```

```python
# Quantidade de cada item
Qsalgadinho = 7
Qdoces = 4
Qsucos = 1
Qrefrigerantes= 2

# Calculo do total
total = (salgadinho * Qsalgadinho) + (Qdoces * doces) + (Qsucos * sucos) + (Qrefrigerantes * refrigerantes)
print("O valor total é: %f" %total)
```

```python
VlrPago = int(input("Informe o valor para calculo de troco: "))
```

```python
#Calculando o troco
troco = VlrPago - total
qntNotas = 0

qntNotasCin = 0
qntNotasVin = 0
qntNotasDez = 0
qntNotasDoi = 0
qntNotasUm = 0

print("O troco é: %f" %troco)

if VlrPago == total:
    print("O valor informado é insuficiênte para pagamento")
else:
    for i in range(len(notas)):
        num_notas = troco // notas[i]

        if num_notas > 0 and notas[i] == 50:
            qntNotasCin = qntNotasCin + 1
        elif num_notas > 0 and notas[i] == 20:
            qntNotasVin = qntNotasVin + 1
        elif num_notas > 0 and notas[i] == 10:
            qntNotasDez = qntNotasDez + 1
        elif num_notas > 0 and notas[i] == 2:
            qntNotasDoi = qntNotasDoi + 1
        elif num_notas > 0 and notas[i] == 1:
            qntNotasUm = qntNotasUm + 1
            
        troco -= num_notas * notas[i]
        qntNotas += num_notas

print("A quantidade minima de notas para o troco é: %d" %qntNotas)
```
