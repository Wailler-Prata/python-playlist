* 1) Ler dois números informados pelo usuário e apresentar o resultado da:
 * a. Soma dos dois números.
 * b. Subtração do primeiro pelo segundo número.
 * c. Multitplicação dos dois números.
 * d. Divisão do primeiro pelo segundo número.

* A)

```python
numero1 = int(input("Informe o primeiro numero: "))
```

```python
numero2 = int(input("Informe o segundo numero: "))
```

```python
"""
soma = numero1 + numero2
print ("Soma: " + str(soma))
"""
soma = numero1 + numero2
print ("Soma: %d" % soma)
```

* B)

```python
subtracao = numero1 - numero2
print("A subtração é: %d" % subtracao)
```

* C)

```python
multipicacao = numero1 * numero2
print("A subtração é: %d" % multipicacao)
```

* D)

```python
divisao = numero1 / numero2
print("A subtração é: %d" % divisao)
```
