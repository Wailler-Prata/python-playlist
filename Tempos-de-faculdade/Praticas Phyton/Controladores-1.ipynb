# Exemplos

# for

```
for x in "Ana Conda":
 print (x)
```

# break

```
for x in "Ana Conda":
 print (x)
 if (x == 'd'):
   break
```

# continue

```
for x in "Ana Conda":
 if (x == 'd'):
   continue
 print (x)
```

# range ()


```
for x in range(5):
 print (x)
```

```
for x in range (-2, 5):
 print (x)
```

```
for x in range (10, 31, 3):
 print (x)
```

# else

```
for x in range (5):
 print (x)
else:
 print ("Fim da estrutura de repetição. \n\tValor final da variável x:", x)
```
