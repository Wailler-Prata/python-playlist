1. Escreva um algoritmo em linguagem de programação Python que leia 3 números a partir
do teclado. O algoritmo deve:
a. Apresentar o maior número.
b. Apresentar o menor número.
c. Apresentar a média aritmética dos três números.

```
numero1 = int(input("Informe o primeiro numero: "))
numero2 = int(input("Informe o segundo numero: "))
numero3 = int(input("Informe o terceiro numero:   "))



# Calcula o maior numero
if numero1 > numero2 and numero1 > numero3:
  print("O maior numero é: %d" %numero1)

elif numero2 > numero1 and numero2 > numero3:
  print("O maior numero é: %d" %numero2)

elif numero3 > numero1 and numero3 > numero2:
  print("O maior numero é: %d" %numero3)



# Calcula o menor numero
if numero1 < numero2 and numero1 < numero3:
  print("O menor numero é: %d" %numero1)

elif numero2 < numero1 and numero2 < numero3:
  print("O menor numero é: %d" %numero2)

elif numero3 < numero1 and numero3 < numero2:
  print("O menor numero é: %d" %numero3)

media = (numero1 + numero1 + numero1) /3

print("A média é: %f" %media)
```

2. Escreva um algoritmo em linguagem de programação Python que leia a coordenada
cartesiana de um ponto. O algoritmo deve imprimir a coordenada do ponto e o quadrante
em que o ponto está situado. Se o ponto estiver na origem, imprimir: "Ponto situado na
origem do plano cartesiano.". Por outro lado, se o ponto estiver situado sobre um dos
eixos, informar se o ponto situa-se na parte positiva ou negativa do eixo X ou Y.

```
numerox = int(input("Informe um numero do eixo x: "))
numeroy = int(input("Informe um numero do eixo y: "))

if numerox < 0 and numeroy < 0:
  print("Ponto: (%d,%d)" %(numerox, numeroy) )
  print("Quadrante 3")
elif numerox < 0 and numeroy > 0:
  print("Ponto: (%d,%d)" %(numerox, numeroy) )
  print("Quadrante 2")
elif numerox > 0 and numeroy > 0:
  print("Ponto: (%d,%d)" %(numerox, numeroy) )
  print("Quadrante 1")
elif numerox > 0 and numeroy < 0:
  print("Ponto: (%d,%d)" %(numerox, numeroy) )
  print("Quadrante 4")
elif numerox == 0 and numeroy == 0:
  print("Ponto situado na origem do plano cartesiano.")
elif numerox < 0 and numeroy == 0:
  print("Ponto: (%d,%d)" %(numerox, numeroy) )
  print("O ponto situa-se na parte negativa do eixo X")
elif numerox == 0 and numeroy < 0:
  print("Ponto: (%d,%d)" %(numerox, numeroy) )
  print("O ponto situa-se na parte negativa do eixo y")
elif numerox > 0 and numeroy == 0:
  print("Ponto: (%d,%d)" %(numerox, numeroy) )
  print("O ponto situa-se na parte positiva do eixo X")
elif numerox == 0 and numeroy > 0:
  print("Ponto: (%d,%d)" %(numerox, numeroy) )
  print("O ponto situa-se na parte positiva do eixo y")
```
