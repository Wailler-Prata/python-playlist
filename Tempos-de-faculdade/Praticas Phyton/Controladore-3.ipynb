#Exemplo de IF ELSE

```
a = int (input ("Informe um número inteiro: "))
b = int (input ("Informe um número inteiro: "))
if a >= b:
 print ("A é maior ou igual a B!")
 print ("Logo, %d >= %d " % (a, b))
else:
 print ("A é menor que B!")
 print ("Logo, %d < %d " % (a, b))
```

#Exemplo de ELIF

```
a = int (input ("Informe um número inteiro: "))
b = int (input ("Informe um número inteiro: "))
if a > b:
 print ("A é maior que B!")
 print ("Logo, %d > %d " % (a, b))
elif a == b:
 print ("A é igual a B!")
 print ("Logo, %d == %d " % (a, b))
else:
 print ("A é menor que B!")
 print ("Logo, %d < %d " % (a, b))
```
