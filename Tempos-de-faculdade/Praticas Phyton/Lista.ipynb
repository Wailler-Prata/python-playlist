# Declaração e Tipo

```
lista = [1, 67.2, True, "Ana Conda"]
print ("Coleção de Dados:", lista)
print ("Tipo da Coleção de Dados:", type (lista))
```

* Outro Exemplo

```
lista = list ()
print ("Coleção de Dados:", lista)
print ("Tipo da Coleção de Dados:", type (lista))
```

* Mais um exemplo

```
lista = list ((1, 67.2, True, "Ana Conda"))
print ("Coleção de Dados:", lista)
print ("Tipo da Coleção de Dados:", type (lista))
```

# Alteração

```
lista = [1, 67.2, True, "Ana Conda"]
print ("Lista:", lista)
lista[1] = 77.5
print ("Nova lista:", lista)
```

# Tamanho da Lista

```
lista = [1, 67.2, True, "Ana Conda"]
print ("Número de itens da lista:", len (lista))
```

# Anexar

```
lista = [1, 67.2, True, "Ana Conda"]
print ("Lista:", lista)
lista.append ("Python")
print ("Nova lista:", lista)
```

# Inserir

```
lista = [1, 67.2, True, "Ana Conda"]
print ("1) Lista:", lista)
lista.insert (0, "Python")
print ("2) Lista:", lista)
lista.insert (2, "Machine Learning")
print ("3) Lista:", lista)
lista.insert (len (lista) - 1, "Data Science")
print ("4) Lista:", lista)
lista.insert (len (lista), "Inteligencia Artificial")
print ("5) Lista:", lista)
```

#Remoção

```
lista = [1, 67.2, True, "Ana Conda"]
print ("Lista:", lista)
lista.remove (67.2)
print ("Nova lista:", lista)
```

# Cópia

```
lista1 = [1, 67.2, True, "Ana Conda"]
lista2 = lista1
print ("Lista1:", lista1)
print ("Lista2:", lista2)
lista1.append ("Python")
print ("Lista1:", lista1)
print ("Lista2:", lista2)
lista2.pop (1)
print ("Lista1:", lista1)
print ("Lista2:", lista2)
```

* Outro exemplo

```
lista1 = [1, 67.2, True, "Ana Conda"]
lista2 = lista1.copy ()
print ("Lista1:", lista1)
print ("Lista2:", lista2)
lista1.append ("Python")
print ("Lista1:", lista1)
print ("Lista2:", lista2)
lista2.pop (1)
print ("Lista1:", lista1)
print ("Lista2:", lista2)
```

* Outro exemplo

```
lista1 = [1, 67.2, True, "Ana Conda"]
lista2 = list (lista1)
print ("Lista1:", lista1)
print ("Lista2:", lista2)
lista1.append ("Python")
print ("Lista1:", lista1)
print ("Lista2:", lista2)
lista2.pop (1)
print ("Lista1:", lista1)
print ("Lista2:", lista2)
```

# Junção

```
lista1 = [1, 67.2]
lista2 = [True, "Ana Conda"]
lista3 = ['Python']
lista = lista1 + lista2 + lista3
print ("Coleção de Dados:", lista)
print ("Tipo da Coleção de Dados:", type (lista))
```

* Outro exemplo

```
lista1 = [1, 67.2]
lista2 = [True, "Ana Conda"]
for x in lista2:
 lista1.append (x)
print ("Lista1:", lista1)
```

* Outro exemplo

```
lista1 = [1, 67.2]
lista2 = [True, "Ana Conda"]
lista1.extend (lista2)
print ("Lista1:", lista1)
```
